```python
df.rename(columns={"Order Id":"order_id","Order Date":"order_date","Ship Mode":"ship_mode","Segment":"segment","Country":"country","City":"city","State":"state","Postal Code":"postal_code","Region":"region","Category":"category","Sub Category":"sub_category","Product Id":"product_id","cost price":"cost_price","List Price":"list_price","Quantity":"quantity","Discount Percent":"discount_percent"},inplace=True)
```

```python
df.columns
```

```python
df.columns=df.columns.str.lower().str.replace(' ','_').str.strip() #standarddize columns name
```

```python
df.columns
```

```python
#CALCULATING DISCOUNTS---derive new columns discount , sale price and profit
df
```

```python
df = df[(df['discount_percent'] >= 0) & (df['discount_percent'] <= 100)] #Validate discount_percent range (0â€“100%)
```

```python
df['profit'] = df['list_price'] - df['cost_price']
# Calculate profit
```

```python
df["discount"]=(df["discount_percent"]/100)*df["list_price"]
df["sale_price"]=df["list_price"]-df["discount"]
df["profit"]=df["sale_price"]-df["cost_price"]
print(df.head())
```

```python
df
```

```python
df["order_date"]=pd.to_datetime(df["order_date"],format="%Y-%m-%d")
```
