```python
df =pd.read_csv("/content/orders.csv")
```

```python
df
```

```python
# data cleaning
df=df.drop(columns=['Unnamed:0'],errors='ignore')
```

```python
df.columns=[col.lower()for col in df.columns]# rename
```

```python
#Using list comprehension to convert spaces to underscores
df.columns = [col.replace(' ', '_') for col in df.columns]

```

```python
df.head()
```

```python
# Check and handle missing values
# Replace missing numerical values with 0 and categorical with 'Unknown'
for column in df.columns:
    if df[column].dtype in ['int64', 'float64']:
        df[column] = df[column].fillna(0)
    else:
        df[column] = df[column].fillna('Unknown')

```

```python
# Ensure 'order_id' column exists and convert to integer
if 'order_id' in df.columns:
    try:
        df['order_id'] = pd.to_numeric(df['order_id'], errors='coerce').fillna(0).astype(int)
    except ValueError:
        print("Error converting 'order_id' to integer. Check for invalid values.")
        raise
else:
    print("Error: 'order_id' column is missing from the dataset.")
    raise KeyError("'order_id' column is required.")
```

```python
# Ensure 'order_id' is an integer for primary/foreign key usage
try:
    df['order_id'] = df['order_id'].astype(int)
except ValueError:
    print("Error converting 'order_id' to integer. Check for invalid values.")
    raise

```

```python
# Verify data types for consistency
print("Data types after cleaning:")
print(df.dtypes)
```

```python
df.info()
```

```python
# Adding new columns in  the Table
#CALCULATING DISCOUNTS---derive new columns discount , sale price and profit
# Calculate Discount (amount of discount)
df['discount'] = (df['list_price'] * df['discount_percent']) / 100



```

```python
# Calculate Sale Price
df['sale_price'] = df['list_price'] - df['discount']

```

```python
# Calculate Profit
df['profit'] = df['sale_price'] - df['cost_price']

```

```python
 # Display the updated DataFrame
df[['list_price', 'discount_percent', 'discount', 'sale_price', 'cost_price', 'profit']].head()
```
